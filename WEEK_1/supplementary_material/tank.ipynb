```python
Q_in = 1    #inflow rate

print('The chosen inflow rate is', Q_in)

A_t = 1      # tank cross-sectional area (m^2)
A_out = 0.1  # outlet area (m^2)

g = 9.81     # gravity

h0 = 0       # initial water level

# Simulation time parameters

T = 100       # total simulation time 
N = 1000  # number of time steps

dt = T / N
```

```python
# outflow rate Q_out

def Q_out(h):
    return A_out * (2 * g * h) ** 0.5
```

```python
print('Outflow flux with h=0.5 is', Q_out(0.5))
```

```python
# lists to store the solution 

times = (N+1)*[0]
h = (N+1)*[0]
h[0] = h0
```

```python
# loop

for i in range(N):
    dh = dt * (Q_in - Q_out(h[i]))
    h[i+1] = h[i] + dh 
    times[i+1] = i*dt
```

```python
# Plot water level over time
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
plt.plot(times, h)
plt.xlabel("Time (s)")
plt.ylabel("Water Level (m)")
plt.title("Water Level Over Time")
plt.grid()
plt.show()
```
